```python
df = pd.read_pickle('/home/patrick/Documents/Alt Coin Proj/data/CombinedData.pkl')
```

```python
df = df.drop(['BTC','USDT','PASC','GNT','GNO'],axis=1)
```

```python
colarr = []
for col in df.columns.unique():
    colarr.append(col[0])
colarr = pd.Series(colarr)
colarr = colarr.unique()
```

```python
def integrity(x):
    print 'Points of Data: %f'  %(len(x))
    index = x.index.values
    time = lambda x: int(str(x).split('T')[1].split(':')[0])
    hour1 = map(time,index)
    
    diff = pd.Series(hour1)
    diff2 = diff.shift(1)
    diff2[0]=diff2[1]-4.0
    delta = abs(diff-diff2)
    delta[delta==20.0] =4.0
    print 'How many time steps missing : %f' %(len(delta[delta<>4.0]))
                         
                         
for cion in colarr:
    print cion
    integrity(df[cion][df[cion]['up'].first_valid_index():df[cion]['up'].last_valid_index()])
```
